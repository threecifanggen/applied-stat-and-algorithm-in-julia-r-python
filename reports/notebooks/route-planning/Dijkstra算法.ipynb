```python
%matplotlib inline
import networkx
import numpy as np
```

本Notebook展示如何获取所有点到"0"的最短距离。

```python
g = networkx.Graph()
g.add_edge("0", "1", length=2)
g.add_edge("0", "2", length=6)
g.add_edge("1", "3", length=5)
g.add_edge("2", "3", length=8)
g.add_edge("3", "5", length=15)
g.add_edge("3", "4", length=10)
g.add_edge("5", "4", length=6)
g.add_edge("5", "6", length=6)
g.add_edge("4", "6", length=2)

networkx.draw(g, with_labels=True)
```

```python
def dijkstra_algo(
    start_node: str,
    g: networkx.Graph):
    unvisited_nodes: set = set(node for node in g.nodes if node != start_node)
    distance_dict: dict = {node: np.Inf if node != '0' else 0 for node in g.nodes}
    
    while True:
        if len(unvisited_nodes) == 0:
            return distance_dict
        else:
            for node in g[start_node]:
                distance_dict[node] = min(distance_dict[node], g[start_node][node]['length'])

```
